```python
import pandas as pd
```

```python
data_df = pd.read_csv('Cities_Weather_Data.csv')
data_df
```

```python
data_df = data_df.drop(columns=['Unnamed: 0'])
data_df
```

```python
data_df= data_df.drop(columns=['id'])
data_df
```

```python
data_df=data_df.set_index('name')
data_df=data_df.sort_index(ascending=True)
data_df
```

```python
html=data_df.to_html()
html_table = open('index.html', 'w')
html_table.write(html)
html_table.close()
```
